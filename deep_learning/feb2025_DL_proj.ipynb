```python
import tensorflow as tf
from tensorflow.keras import layers, Model, Input
from tensorflow.keras.layers import Conv2D, Conv2DTranspose, MaxPooling2D, BatchNormalization
from tensorflow.keras.layers import Reshape, concatenate, Activation, Dense, Add, Dropout, LayerNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import numpy as np
from tensorflow.keras.datasets import mnist, fashion_mnist
import matplotlib.pyplot as plt
import tensorflow.keras.backend as K

# ----------------------------
# Data visualization function (with more modification to draw predicted and ground truth)
# ----------------------------
def show_images(x, y1, y2, pred1=None, pred2=None):
    fig, ax = plt.subplots(1, 5, figsize=(20,4))
    ax[0].imshow(x.squeeze(), cmap='gray')
    ax[0].set_title('Input (Combined)')
    ax[0].axis('off')
    ax[1].imshow(y1.squeeze(), cmap='gray')
    ax[1].set_title('Ground Truth MNIST')
    ax[1].axis('off')
    ax[2].imshow(y2.squeeze(), cmap='gray')
    ax[2].set_title('Ground Truth Fashion')
    ax[2].axis('off')
    if pred1 is not None:
        ax[3].imshow(pred1.squeeze(), cmap='gray')
        ax[3].set_title('Predicted MNIST')
        ax[3].axis('off')
    if pred2 is not None:
        ax[4].imshow(pred2.squeeze(), cmap='gray')
        ax[4].set_title('Predicted Fashion')
        ax[4].axis('off')
    plt.show()

best_batch_size = 256

# ----------------------------
# Load and preprocess data
# ----------------------------
(mnist_x_train, _), (mnist_x_test, _) = mnist.load_data()
(fashion_x_train, _), (fashion_x_test, _) = fashion_mnist.load_data()

mnist_x_train = np.pad(mnist_x_train, ((0,0),(2,2),(2,2)), mode='constant') / 255.
mnist_x_test  = np.pad(mnist_x_test,  ((0,0),(2,2),(2,2)), mode='constant') / 255.
fashion_x_train = np.pad(fashion_x_train, ((0,0),(2,2),(2,2)), mode='constant') / 255.
fashion_x_test  = np.pad(fashion_x_test,  ((0,0),(2,2),(2,2)), mode='constant') / 255.

mnist_x_train = np.expand_dims(mnist_x_train, -1)
mnist_x_test  = np.expand_dims(mnist_x_test, -1)
fashion_x_train = np.expand_dims(fashion_x_train, -1)
fashion_x_test  = np.expand_dims(fashion_x_test, -1)

# ----------------------------
# Data generator (as defined in the docs)
# ----------------------------
def datagenerator(x1, x2, batchsize):
    n1 = x1.shape[0]
    n2 = x2.shape[0]
    while True:
        num1 = np.random.randint(0, n1, batchsize)
        num2 = np.random.randint(0, n2, batchsize)
        x_data = (x1[num1] + x2[num2]) / 2.0
        y_data = (x1[num1], x2[num2])
        yield x_data, y_data

# ----------------------------
# 1) Define the CBAM module (Convolutional Block Attention Module)
# ----------------------------
def cbam_block(input_feature, ratio=16):
    channel = input_feature.shape[-1]

    shared_layer_one = Dense(channel // ratio,
                             activation='relu',
                             kernel_initializer='he_normal',
                             use_bias=True,
                             bias_initializer='zeros')
    shared_layer_two = Dense(channel,
                             kernel_initializer='he_normal',
                             use_bias=True,
                             bias_initializer='zeros')

    avg_pool = tf.keras.layers.GlobalAveragePooling2D()(input_feature)
    avg_pool = Reshape((1, 1, channel))(avg_pool)
    avg_pool = shared_layer_one(avg_pool)
    avg_pool = shared_layer_two(avg_pool)

    max_pool = tf.keras.layers.GlobalMaxPooling2D()(input_feature)
    max_pool = Reshape((1, 1, channel))(max_pool)
    max_pool = shared_layer_one(max_pool)
    max_pool = shared_layer_two(max_pool)

    cbam_feature = Add()([avg_pool, max_pool])
    cbam_feature = Activation('sigmoid')(cbam_feature)
    channel_refined_feature = tf.keras.layers.Multiply()([input_feature, cbam_feature])

    avg_pool_spatial = tf.keras.layers.Lambda(lambda x: K.mean(x, axis=3, keepdims=True))(channel_refined_feature)
    max_pool_spatial = tf.keras.layers.Lambda(lambda x: K.max(x, axis=3, keepdims=True))(channel_refined_feature)
    concat = concatenate([avg_pool_spatial, max_pool_spatial], axis=3)
    spatial_attention = Conv2D(filters=1, kernel_size=(7,7), padding='same',
                               activation='sigmoid', kernel_initializer='he_normal',
                               use_bias=False)(concat)
    refined_feature = tf.keras.layers.Multiply()([channel_refined_feature, spatial_attention])

    return refined_feature

# ----------------------------
# 2) Define the standard Transformer Encoder Block (Bottleneck) (use-case is after encoder layer)
# ----------------------------
def transformer_encoder_block(inputs, num_heads=4, embed_dim=192, ff_dim=640,
                              dropout_rate=0.1, epsilon=1e-6):
    x_norm1 = LayerNormalization(epsilon=epsilon)(inputs)
    attn_output = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)(x_norm1, x_norm1)
    attn_output = Dropout(dropout_rate)(attn_output)
    out1 = Add()([inputs, attn_output])

    x_norm2 = LayerNormalization(epsilon=epsilon)(out1)
    ffn = Dense(ff_dim, activation='relu')(x_norm2)
    ffn = Dense(embed_dim)(ffn)
    ffn = Dropout(dropout_rate)(ffn)
    out2 = Add()([out1, ffn])

    return out2

# ----------------------------
# 3) Define a Lightweight Transformer Block for the Decoder Branch
# ----------------------------
def lightweight_transformer_block(inputs, num_heads=2, embed_dim=64, ff_dim=128,
                                  dropout_rate=0.1, epsilon=1e-6):
    x_norm1 = LayerNormalization(epsilon=epsilon)(inputs)
    attn_output = layers.MultiHeadAttention(num_heads=num_heads, key_dim=embed_dim)(x_norm1, x_norm1)
    attn_output = Dropout(dropout_rate)(attn_output)
    out1 = Add()([inputs, attn_output])

    x_norm2 = LayerNormalization(epsilon=epsilon)(out1)
    ffn = Dense(ff_dim, activation='relu')(x_norm2)
    ffn = Dense(embed_dim)(ffn)
    ffn = Dropout(dropout_rate)(ffn)
    out2 = Add()([out1, ffn])

    return out2

# ----------------------------
# 4) Build Model:
# Decoder with an Autoencoding (UNet) Sub‐Structure + Embedded Lightweight Transformer & Attention.
# ----------------------------
def build_model2(input_shape=(32, 32, 1),
                 num_transformer_blocks=3,
                 embed_dim=192,
                 ff_dim=640,
                 num_heads=4,
                 dropout_rate=0.1,
                 epsilon=1e-6):
    # --- Encoder ---
    input_img = Input(shape=input_shape, name='input_img')

    # Encoder Block 1
    e1 = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
    e1 = Conv2D(32, (3, 3), activation='relu', padding='same')(e1)
    p1 = MaxPooling2D((2, 2), padding='same')(e1)

    # Encoder Block 2
    e2 = Conv2D(64, (3, 3), activation='relu', padding='same')(p1)
    e2 = Conv2D(64, (3, 3), activation='relu', padding='same')(e2)
    p2 = MaxPooling2D((2, 2), padding='same')(e2)

    # Encoder Block 3
    e3 = Conv2D(128, (3, 3), activation='relu', padding='same')(p2)
    e3 = Conv2D(128, (3, 3), activation='relu', padding='same')(e3)
    encoded = MaxPooling2D((2, 2), padding='same')(e3)

    # --- Transformer Bottleneck ---
    x = Reshape((4 * 4, 128))(encoded)
    x = Dense(embed_dim)(x)
    pos_embedding = tf.Variable(tf.random.normal([1, 16, embed_dim]), trainable=True, name='pos_embedding')
    x = x + pos_embedding
    for _ in range(num_transformer_blocks):
        x = transformer_encoder_block(x, num_heads=num_heads, embed_dim=embed_dim,
                                      ff_dim=ff_dim, dropout_rate=dropout_rate, epsilon=epsilon)
    encoded_trans = Reshape((4, 4, embed_dim))(x)

    # ----------------------------
    # Decoder Branch 1 (let's say MNIST generator)
    # ----------------------------
    # Stage 1: Initial Upsampling
    d1 = Conv2DTranspose(128, (3, 3), strides=(2, 2),
                           activation='relu', padding='same')(encoded_trans)
    d1 = BatchNormalization()(d1)
    d1 = concatenate([d1, e3])
    d1 = Conv2D(128, (3, 3), activation='relu', padding='same')(d1)
    d1 = BatchNormalization()(d1)
    d1 = cbam_block(d1)

    d1 = Conv2DTranspose(64, (3, 3), strides=(2, 2),
                           activation='relu', padding='same')(d1)
    d1 = BatchNormalization()(d1)
    d1 = concatenate([d1, e2])
    d1 = Conv2D(64, (3, 3), activation='relu', padding='same')(d1)
    d1 = BatchNormalization()(d1)
    d1 = cbam_block(d1)
    # Now d1 has shape ~ (16, 16, 64)

    # Stage 2: Autoencoder Branch with Lightweight Transformer
    # Downsampling (to ~8x8)
    x_auto = Conv2D(64, (3, 3), strides=2,
                    activation='relu', padding='same')(d1)  # (8,8,64)
    x_auto = BatchNormalization()(x_auto)

    # Reshape to sequence for transformer processing
    b, h, w, c = K.int_shape(x_auto)
    seq_len = h * w  # 8*8 = 64
    x_auto_seq = Reshape((seq_len, c))(x_auto)

    # Apply a lightweight transformer layer (can stack more than one if desired)
    x_auto_seq = lightweight_transformer_block(x_auto_seq, num_heads=2, embed_dim=64,
                                               ff_dim=128, dropout_rate=dropout_rate, epsilon=epsilon)

    # Reshape back to spatial feature map
    x_auto = Reshape((h, w, c))(x_auto_seq)

    # Upsample back to 16x16
    x_auto = Conv2DTranspose(64, (3, 3), strides=2,
                              activation='relu', padding='same')(x_auto)
    x_auto = BatchNormalization()(x_auto)

    # Use the refined features from the autoencoder branch
    d1 = x_auto

    # Stage 3: Final Upsampling
    d1 = Conv2DTranspose(32, (3, 3), strides=(2, 2),
                           activation='relu', padding='same')(d1)
    d1 = BatchNormalization()(d1)
    d1 = concatenate([d1, e1])
    d1 = Conv2D(32, (3, 3), activation='relu', padding='same')(d1)
    d1 = BatchNormalization()(d1)
    d1 = cbam_block(d1)

    decoded1 = Conv2D(1, (3, 3), activation='sigmoid',
                      padding='same', name='decoded1')(d1)

    # ----------------------------
    # Decoder Branch 2 (duplicate branch; let's say FASHION-MNIST generator)
    # ----------------------------
    d2 = Conv2DTranspose(128, (3, 3), strides=(2, 2),
                           activation='relu', padding='same')(encoded_trans)
    d2 = BatchNormalization()(d2)
    d2 = concatenate([d2, e3])
    d2 = Conv2D(128, (3, 3), activation='relu', padding='same')(d2)
    d2 = BatchNormalization()(d2)
    d2 = cbam_block(d2)

    d2 = Conv2DTranspose(64, (3, 3), strides=(2, 2),
                           activation='relu', padding='same')(d2)
    d2 = BatchNormalization()(d2)
    d2 = concatenate([d2, e2])
    d2 = Conv2D(64, (3, 3), activation='relu', padding='same')(d2)
    d2 = BatchNormalization()(d2)
    d2 = cbam_block(d2)

    x_auto2 = Conv2D(64, (3, 3), strides=2,
                     activation='relu', padding='same')(d2)  # (8,8,64)
    x_auto2 = BatchNormalization()(x_auto2)
    b2, h2, w2, c2 = K.int_shape(x_auto2)
    seq_len2 = h2 * w2
    x_auto2_seq = Reshape((seq_len2, c2))(x_auto2)
    x_auto2_seq = lightweight_transformer_block(x_auto2_seq, num_heads=2, embed_dim=64,
                                                ff_dim=128, dropout_rate=dropout_rate, epsilon=epsilon)
    x_auto2 = Reshape((h2, w2, c2))(x_auto2_seq)
    x_auto2 = Conv2DTranspose(64, (3, 3), strides=2,
                              activation='relu', padding='same')(x_auto2)  # back to (16,16,64)
    x_auto2 = BatchNormalization()(x_auto2)

    d2 = x_auto2

    d2 = Conv2DTranspose(32, (3, 3), strides=(2, 2),
                           activation='relu', padding='same')(d2)
    d2 = BatchNormalization()(d2)
    d2 = concatenate([d2, e1])
    d2 = Conv2D(32, (3, 3), activation='relu', padding='same')(d2)
    d2 = BatchNormalization()(d2)
    d2 = cbam_block(d2)

    decoded2 = Conv2D(1, (3, 3), activation='sigmoid',
                      padding='same', name='decoded2')(d2)

    model = Model(inputs=input_img, outputs=[decoded1, decoded2],
                  name='Model2_UNet_Autoencoder_Transformer')
    return model

# ----------------------------
# 5) Start and compile Model 2
# ----------------------------
model2 = build_model2()
model2.summary()
model2.compile(optimizer='adam', loss='mse')

# ----------------------------
# Custom Callback to store best model weights in memory (since I had some problem with Model_checkpoint i implement mine)
# ----------------------------
class BestModelSaver(tf.keras.callbacks.Callback):
    def __init__(self, monitor='loss', mode='min'):
        super(BestModelSaver, self).__init__()
        self.monitor = monitor
        self.mode = mode
        if self.mode == 'min':
            self.best = np.Inf
        else:
            self.best = -np.Inf
        self.best_weights = None

    def on_epoch_end(self, epoch, logs=None):
        current = logs.get(self.monitor)
        if current is None:
            return
        if self.mode == 'min':
            if current < self.best:
                self.best = current
                self.best_weights = self.model.get_weights()
                print("Epoch {}: New best {}: {}".format(epoch, self.monitor, current))
        else:
            if current > self.best:
                self.best = current
                self.best_weights = self.model.get_weights()
                print("Epoch {}: New best {}: {}".format(epoch, self.monitor, current))

best_model_saver = BestModelSaver(monitor='loss', mode='min')

# ----------------------------
# 6) Set up training with callbacks
# ----------------------------
steps_per_epoch = 1000
epochs = 80

early_stopping = EarlyStopping(monitor='loss', patience=5, restore_best_weights=False)
reduce_lr = ReduceLROnPlateau(monitor='loss', factor=0.5, patience=5, min_lr=1e-6, verbose=1)

callbacks = [early_stopping, reduce_lr, best_model_saver]

train_gen = datagenerator(mnist_x_train, fashion_x_train, best_batch_size)
history = model2.fit(train_gen,
                     steps_per_epoch=steps_per_epoch,
                     epochs=epochs,
                     callbacks=callbacks,
                     verbose=1)

# ----------------------------
# 7) After training, update model with the best weights stored in memory
# ----------------------------
if best_model_saver.best_weights is not None:
    model2.set_weights(best_model_saver.best_weights)
    print("Model updated with best weights from training.")
else:
    print("No best weights found; using current model weights.")

# Assign the best model to a variable (here, best_model is simply model2 with the best weights)
best_model = model2

# ----------------------------
# 8) Define evaluation function
# ----------------------------
def eval_model(model, test_gen, num_evals=10):
    eval_results = []
    for i in range(num_evals):
        x_eval, (y_eval1, y_eval2) = next(test_gen)
        pred1, pred2 = model.predict(x_eval)
        mse1 = np.mean((pred1 - y_eval1)**2)
        mse2 = np.mean((pred2 - y_eval2)**2)
        mse = (mse1 + mse2) / 2.0
        eval_results.append(mse)
        print(f"Evaluation {i+1}: MSE MNIST = {mse1:.5f}, MSE Fashion = {mse2:.5f}, Combined MSE = {mse:.5f}")
    print("\nOverall evaluation:")
    print("Mean MSE =", np.mean(eval_results))
    print("Std MSE =", np.std(eval_results))
    return eval_results

test_gen_large = datagenerator(mnist_x_test, fashion_x_test, 5000)

# ----------------------------
# 9) Evaluate the best model on the test set
# ----------------------------
print("Evaluating best model on test set:")
eval_results = eval_model(best_model, test_gen_large, num_evals=10)

# ----------------------------
# 10) Visualize some predictions using the best model
# ----------------------------
vis_gen = datagenerator(mnist_x_test, fashion_x_test, 1)
x_vis, (y_vis1, y_vis2) = next(vis_gen)
pred_vis1, pred_vis2 = best_model.predict(x_vis)
show_images(x_vis[0], y_vis1[0], y_vis2[0], pred_vis1[0], pred_vis2[0])
```

# Model Description
  1. Encoder: CNN-Based Downsampling (UNet)

  A UNet-style encoder with convolutional layers and max-pooling progressively downsamples the input while extracting spatial features.
  The feature map is reshaped for transformer processing.

  2. Transformer-Based Bottleneck

  The encoder output is passed through multiple transformer blocks with multi-head self-attention, layer normalization, and feed-forward networks to model long-range dependencies.
  Residual connections are used to enhance gradient flow.

  3. Dual Decoder: Two Parallel Decoders

  Each decoder consists of:
    Upsampling (UNet-style): Gradually reconstructs spatial features. Which complete encoder layers.

    Autoencoder Blocks: Each upsampling step is followed by another downsampling-then-upsampling block to refine feature representation.
    Attention Modules: Self-attention layers further enhance feature dependencies.
    Channel and Spatial Attention Mechanisms: Multiplicative attention mechanisms refine feature selection.

  4. Final Output

  Two independent decoder heads produce reconstructed outputs (decoded1 and decoded2), allowing different perspectives on feature extraction.

**Why this structure? Experience**

My base model was a U-Net structure with a transformer-based bottleneck, which achieved a Mean MSE of 0.00056. Upon visualizing the output, I observed that the model was effective at separating images and preserving the general shape. However, it lacked accuracy in finer details across different locations.

The first model I tried after this was transfer learning from EMNIST. I experimented with freezing encoder layers, freezing transformers, and gradual freezing. The results ranged from a Mean MSE of 0.00065 to 0.0010.

Next, I used a pretrained Autoencoder. In one configuration, the Autoencoder was trained for 50 epochs on the base model's output to learn the distribution of reconstruction errors, then integrated back into the model for another 20 epochs. In another configuration, I froze the base model encoder and followed the same process. Both approaches achieved a Mean MSE of no better than 0.00057.

I then added attention blocks to the decoder, but this did not significantly improve the results (Mean MSE = 0.0005678).

Realizing the need for better reconstruction, I added U-Net reconstruction layers to each decoder branch, thinking the additional complexity might be worthwhile. This structure achieved the best Mean MSE across all my models.

### Although this structure is unconventional, as I could not find any prior work supporting it, it produced the best results in my experiments.

Critique: A configuration I wanted to test but couldn’t was to create a dataset using 1% of data from one class and 99% from another. The target would be the 99% class data. Hypothetically, training an Autoencoder on this dataset could improve the baseline Mean MSE of 0.00056. After training the baseline, this new Autoencoder could be used to remove the less than 1% (0.01) of information from the final output.
